```python
file='/Users/vijayasri/ml/EasyVisa.csv'
wage=pd.read_csv(file)
wage_data=wage['prevailing_wage']
mean=np.mean(wage_data)
median=np.median(wage_data)
maxx=np.max(wage_data)
minn=np.min(wage_data)
std=np.std(wage_data)
per=np.percentile(wage_data,25)
per50=np.percentile(wage_data,50)
per75=np.percentile(wage_data,75)
quar=np.percentile(wage_data,0.25)
list=[mean,median,maxx,minn,std,per,per50,per75,quar]
rows=['mean','median','maxx','minn','std','plt','per50','per75','quar']
pd.DataFrame(list,columns=['prevailing_wage'],index=[rows])
```

```python
wage_data=wage['prevailing_wage']
wage_data_25p=len(wage_data[wage_data<np.percentile(wage_data,25)])
count_50p=len(wage_data[wage_data<np.percentile(wage_data,50)])
count_75p=len(wage_data[wage_data<np.percentile(wage_data,75)])
columns=['wage_data_25p','count_50p','count_75p']
list=[wage_data_25p,count_50p,count_75p]

s1=pd.DataFrame(zip(columns,list),columns=['count','till'])
plt.bar('count','till',data=s1)
```
