```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
file='/Users/vijayasri/ML/EasyVisa.csv'
visa_df=pd.read_csv(file)
val=visa_df['prevailing_wage']
mean_val=np.mean(visa_df['prevailing_wage'])
std1=np.std(visa_df['prevailing_wage'])
num=visa_df['prevailing_wage']-mean_val
out=num/std1
visa_df['prevailing_wage_zscore']=out
visa_df
```
