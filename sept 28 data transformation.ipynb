```python
data=np.random.exponential(size=1000)
data1=np.log(data)
plt.hist(data1,bins=10,label='data exp')
plt.legend()
plt.show()
```

```python
data=np.random.exponential(size=10000)
exp=1/data
plt.hist(exp,bins=10,label='exp')
plt.legend()
plt.show()
```

```python
sqr=np.sqrt(data)
plt.hist(sqr,label='sqrplt')
plt.legend()
plt.figure(figsize=(10,10))
plt.show()
```
