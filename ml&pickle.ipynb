```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
import sklearn 
import warnings
warnings.filterwarnings('ignore')
```

```python
file='/Users/vijayasri/ML/winequality_red.csv'
main=pd.read_csv(file)
main.isnull().sum()
main.reset_index(drop=True,inplace=True)
X=main.drop('quality',axis=1)
y=main['quality']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=130)
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
y_prediction=LR.predict(X_test)
from sklearn.metrics import r2_score,mean_squared_error
mms=mean_squared_error(y_test,y_prediction)
r2=r2_score(y_test,y_prediction)
X=main.drop('quality',axis=1)
```

```python
import pickle
pickle.dump(LR,open('linear_wine.pkl','wb'))
model=pickle.load(open('linear_wine.pkl','rb'))
```
